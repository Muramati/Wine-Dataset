## Using the  Wine Quality Dataset, test the claim that Wines with a higher quality rating have a higher median alcohol content than wines with a lower quality rating.

### 1.  Formulating the Hypothesis
Null Hypothesis:The median alcohol content is the same regardless of the quality.

Alternative Hypothesis: Wine with higher quality rating has higher median alcohol content compared to wine lower in the quality rating.



 ### 2. Confidence Level

The confidence level is set at 95% therefore making the alpha 5%.

Î± = 0.05

### 3. Code

```python
# Importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats 
```

```python
# Load csv to df
wine = pd.read_csv('winequality-red.csv', sep=';', header=0)
wine.head(15)
```

```python
# get unique items in the quality column

wine.quality.unique()
```

```python
# group by quality and finding median of alcohol content

winequality = wine.groupby('quality')['alcohol'].median().to_dict()

winequality
```

```python
# loop over winequality items and sort wine in lower quality and higher quality list

lower_quality = []
high_quality =  []
for key,value in  winequality.items():
    if key <= 5:
        lower_quality.append(value)
    else:
        high_quality.append(value)
```

```python
#print the results.
print ( 'high quality : ' ,high_quality)
print ( 'lower quality : ' ,lower_quality)
```

```python
#  performing a two sample t test

def two_sample_test(sample_1, sample_2):
    results = stats.ttest_ind(a = sample_1, b = sample_2)

    return results


z_score, p_value = two_sample_test(lower_quality, high_quality)
```

### Conclusion:


```python
if (p_value < 0.05):
    print(f"Reject null hypothesis with {100-(p_value*100)}% confidence level")
else:
    print("Fail to reject null hypothesis")
```
